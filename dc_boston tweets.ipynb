```python
import numpy as np
import re
import pandas as pd
import seaborn as sns
from wordcloud import WordCloud, STOPWORDS
from matplotlib import pyplot as plt
```

```python
data_set = pd.read_csv('COVID-tweets.csv')
```

```python
data_set_DC = data_set[data_set['Tweet Location'] == 'Washington, DC'].reset_index()
```

```python
data_set_DC.drop('index', axis = 1, inplace = True)
```

```python
data_set_DC
```

```python
data_set_boston = data_set[data_set['Tweet Location'] == 'Boston'].reset_index()
```

```python
data_set_boston.drop('index', axis = 1, inplace = True)
```

```python
data_set_boston
```

```python
x = pd.merge(data_set_DC, data_set_boston, how = 'outer')
```
