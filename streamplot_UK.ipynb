```python
import numpy as np
import re
import pandas as pd
import seaborn as sns
from wordcloud import WordCloud, STOPWORDS
from matplotlib import pyplot as plt
```

```python
data_set = pd.read_csv('COVID-tweets.csv')
```

```python
dates = data_set['Tweet Posted Time (UTC)']
```

```python
count = 0
for i in dates:
    #print(i)
    #print(type(i))
    if "Feb" in i:
        #print(True)
        count+=1
data_set_feb = data_set.iloc[:count,0:].copy()

data_set_feb
```

```python
data_set_uk = data_set_feb[(data_set_feb['Tweet Location'] == 'UK') | (data_set_feb['Tweet Location'] == 'London, UK')].reset_index()
```

```python
data_set_uk.drop('index', inplace = True, axis = 1)
```

```python
data_set_uk
```

```python
#data_set_uk['Tweet Location'].rename == 'UK', axis = 1, inplace = True]
```

```python
new = data_set_uk['Tweet Posted Time (UTC)'].str.split(" ", n = 1, expand = True) 
  
# making separate first name column from new data frame 
data_set_uk['Date'] = pd.to_numeric(new[0]) 
  
# making separate last name column from new data frame 
data_set_uk["month_time"] = new[1]
data_set_uk.drop(columns = ['Tweet Posted Time (UTC)'], inplace = True)
```

```python
data_set_uk
```

```python
data_set_uk2 = data_set_uk[['Date', 'month_time','Tweet Content', 'Tweet Location']]
```

```python
data_set_uk2
```

```python
data_set_uk2.loc[3,'Tweet Content']
```

```python
m = pd.DataFrame()
#k = 0
for i in range(9, 29):
    n = " "
    for j in range(107):
        if i == data_set_uk2.loc[j,'Date']:
            #print('Loop entered', i)
            print(data_set_uk2.loc[j,'Tweet Content'])
            n += " " + data_set_uk2.loc[j,'Tweet Content']
    m.loc[0, 'Day ' + str(i)] = n
```

```python
import nltk

from nltk.corpus import stopwords

nltk.download('stopwords')
```

```python
stop_words = ["https", "co", "de", "The","for", "For", "this", "still", "in", "U", "9", "RT", "the","a", "q", "says" "OUR","Our" ,"our", "This", "In", "1"] + list(stopwords.words('english'))
```

```python
y9 = [] 
res = re.findall(r'\w+', str(x9))
for i in range(len(res)):
    if (str(res[i]) == 'https') | (str(res[i]) == 'co'):
        pass
    else:
        y9.append(str(res[i]))

        


z9 = [word.lower() for word in y9 if word not in stop_words]

import collections
counts_no_urls = collections.Counter(z9)
counts_no_urls.most_common(30)
```
