```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
k1 = []
for i in range(1,10):
    k1.append(str(i))
```

```python
stop_words = k1 + ["https","one","due",'go','08','http',"www","The","for", "For", "la","de","es",'en', "el","com", "still","pic","twitter", "This", "In", "1"] + list(stopwords.words('english'))
```

```python
date_files_08 = ['tweets_2020-08-01.csv','tweets_2020-08-02.csv','tweets_2020-08-03.csv','tweets_2020-08-04.csv',
              'tweets_2020-08-05.csv','tweets_2020-08-06.csv','tweets_2020-08-07.csv','tweets_2020-08-08.csv',
              'tweets_2020-08-09.csv','tweets_2020-08-10.csv','tweets_2020-08-11.csv','tweets_2020-08-12.csv',
              'tweets_2020-08-13.csv','tweets_2020-08-14.csv','tweets_2020-08-15.csv','tweets_2020-08-16.csv',
              'tweets_2020-08-17.csv','tweets_2020-08-18.csv','tweets_2020-08-19.csv','tweets_2020-08-20.csv',
              'tweets_2020-08-21.csv','tweets_2020-08-22.csv', 'tweets_2020-08-23.csv','tweets_2020-08-24.csv']
```

```python
dates_dict_08 = {}

for i in range(len(date_files_08)):
    data_set = pd.read_csv(date_files_08[i])
    data_set2 = data_set[['date', 'tweet','hashtags']]
    
    n = ""
    for j in range(len(data_set2)):
        n += " " + data_set2.loc[j,'tweet'] 
        
    y = [] 
    res = re.findall(r'\w+', str(n))
    for k in range(len(res)):
        if (str(res[k]) == 'https'):
            pass
        else:
            y.append(str(res[k].lower()))
            
    z = [word for word in y if word not in stop_words]
    counts_no_urls = collections.Counter(z)
    a1 = counts_no_urls.most_common(80)
    dates_dict_08.update({date_files_08[i][7:17] : a1})
```

```python
dates_dict_08
```

```python
p_count = []
p = []
for i in dates_dict_08.values():
    words = []
    words_count = []
    print(i)
    for k in range(80):
        words.append(i[k][0])
        words_count.append((i[k][0], i[k][1]))
    p.append(words)
    p_count.append(words_count)
```

```python
l = []
for i in p[0]:
    print(i)
    count = 0
    for j in range(0,24):
        if i in p[j]:
            count += 1
            #print("True, Set", j)
    l.append((i,count))
```

```python
req_words2 = sorted(l, key=lambda x: x[1],reverse = True)
```

```python
req_words = req_words2[:25]
```

```python
out_words = [i[0] for i in req_words]
```

```python
out_words
```

```python
out = []
for i in out_words:
    temp = []
    print(i)
    for k in range(0,24):
        if i in [j[0] for j in p_count[k]]:
            p1 = [j[1] for j in p_count[k] if j[0] == i]
            temp.append(p1[0])
        else:
            temp.append(0)
    
    out.append((i, temp))
    
```

```python
for i in range(len(out)):
    print(out[i])
```
